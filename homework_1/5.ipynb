```python
import scipy.stats as stats
import matplotlib.pyplot as plot
import numpy
%matplotlib inline
```

```python
N = 1000
samples = 5000
E = 15.75
D = 0.24
bootstrap = [stats.norm.rvs(E, scale=numpy.sqrt(D), size=N) for i in range(samples)]
```

```python
statistics = [
    [(numpy.sum(sample[:i]) - i * E) / (numpy.sqrt(i * D)) for i in range(1, N + 1)]
    for sample in bootstrap
]
```

```python
for i in range(5):
    plot.figure()
    plot.plot(statistics[i])
```

Построенные статистики как числове последовательности не сходятся, однако вероятность, что на какой-то выборке статистика будет лежать в малой окрестности нуля начиная с какого-то $k$ ненулевая. То есть выборка, на которой статистика как числовая последовательность сходится существует.

```python
values = [statistic[-1] for statistic in statistics] 
```

```python
figure, axis = plot.subplots()

axis.hist(values, 50, normed=1, color="green", alpha=0.6)
points = numpy.linspace(-4, 4, 1000)
axis.plot(points, (lambda x: stats.norm.pdf(x, 0, scale=1))(points), color="black")
```

Так как распределение выборок нормальное, значения статистики $T_k$ для разных выборок распределены нормально с нулевым средним и единичной дисперсией для любого $k\le n$ (следует из ствойств нормального распределения). Неудивительно, что распределение конечных точек статистик для всех выборок имеет именно такое распределение (что подтверждается графиком). ЦПТ выполняется.
