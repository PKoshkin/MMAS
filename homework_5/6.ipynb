```python
import pandas
from scipy import stats
import numpy
```

```python
data = pandas.read_csv("6.csv")
```

```python
data
```

```python
columns = ['Water', 'SP', 'Fine Aggr.', 'SLUMP(cm)', 'FLOW(cm)', 'Compressive Strength (28-day)(Mpa)']
```

```python
train_samples = numpy.array([data[column][:int(len(data) / 2)] for column in columns])
test_samples = numpy.array([data[column][int(len(data) / 2) + 1:] for column in columns])
```

Как здесь использовать коэффициенты корреляции? Можно с помощью них оценить матрицу ковариации и таким образом оценить параметры гауссовского распределения. Затем сгенерировать бутстреп и проверить, что выборки из него не сильно отличаются от данной. Однако коэффициенты корреляции нормированы. Так что лучше просто взять выборочные ковариации.

```python
cov = numpy.cov(train_samples)
E = [numpy.mean(train_samples) for sample in train_samples]
```

```python
# Сгенерируем выборку с оцененными параметрами, от которой будем считать отклонение
test = stats.multivariate_normal.rvs(E, cov, size=len(train_samples[0])).T
```

```python
# Считаем среднее отклонение от изначально снегерированой выборки для еще кучи выборок
errors = []
for i in range(1000):
    new_samples = stats.multivariate_normal.rvs(E, cov, size=len(train_samples[0])).T
    errors.append(numpy.mean([(x - y) ** 2 for x, y in zip(test, new_samples)]))
print("mean error:", numpy.mean(errors))
print("max error:", numpy.max(errors))
```

```python
numpy.mean([(x-y)**2 for x, y in zip(test_samples, test)])
```

Отклонение на тестовой части получилось больше многократного "максимумального порогового" отклонения. Гипотезу о нормальности можно отвергнуть.
