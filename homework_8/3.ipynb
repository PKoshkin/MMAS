```r
# Генерируем данные
norm_sample <- rnorm(n = 500, mean = 1, sd = 1)
```

```r
# Построим гистограмму, чтобы проверить, что получилось то, что хотели.
hist(norm_sample)
```

```r
# Получаем и выводим оценки параметров
mu = mean(norm_sample)
sigma = sqrt(var(norm_sample))
print(mu)
print(sigma)
```

```r
# Считаем квантили и получаем значения, лежащие в нужном интервале
alpha = 0.05
u_r = qnorm(mean = 1, sd = 1, 1 - alpha / 2)
u_l = qnorm(mean = 1, sd = 1, alpha / 2)
values = norm_sample[(norm_sample > u_r) | (norm_sample < u_l)]
```

```r
print(length(values))
print(u_r)
print(u_l)
```
