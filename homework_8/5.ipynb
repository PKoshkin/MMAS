```r
# Определяем функции из условия
f <- function(x) {
    ifelse((x <= 1) & (x >= 0), 20 * x * (1 - x) ^ 3, 0) 
}
g <- function(x) {
    ifelse((x <= 1) & (x >= 0), 1, 0) 
}
c <- 20
```

```r
# Пишем, функцию - генератор
generate <- function() {
    repeat {
        Y <- runif(n=1)
        t <- runif(n=1)
        if (t <= f(Y) / (c * g(Y))) {
            return(Y)
        }
    }
}
```

```r
# Генерируем выборку нужного размера
N = 1000
sample = generate()
for (i in 2:N) {
    sample = c(sample, generate())
}
```

```r
# Построим и сравним график плотности и гистограмму сгенерированных данных
plot(f, 0, 1)
hist(sample)
```
