```r
data <- read.table('3.txt', header = FALSE)
```

```r
data
```

```r
n <- 400
q <- 100
```

```r
train_data = data[1:n, 1:ncol(data)]
test_data = data[(n + 1):(n + q), 1:ncol(data)]
```

С помощью разных методов bestglm посмотрим варианты лучших моделей.

```r
library('bestglm')
for (ic in c("AIC", "BIC", "BICg", "BICq", "LOOCV", "CV")) {
    print(bestglm(train_data, IC=ic))
}
```

Есть 2 варианты. Оставить все признаки, или все кроме 4.

```r
model_1 <- lm(formula = V6 ~ V1 + V2 + V3 + V5, data = train_data)
model_2 <- lm(formula = V6 ~ V1 + V2 + V3 + V4 + V5, data = train_data)
```

Проверим, какова мультиколлинеарность этих двух моделей.

```r
XX = t(data.matrix(train_data[1:n,1:5])) %*% data.matrix(train_data[1:n,1:5])
lambdas = eigen(XX)$values
print(max(lambdas) / min(lambdas))
```

```r
XX = t(data.matrix(train_data[1:n,c(1:3,5)])) %*% data.matrix(train_data[1:n,c(1:3,5)])
lambdas = eigen(XX)$values
print(max(lambdas) / min(lambdas))
```

Оказалось, что в модели со всеми столбцами есть почти линейно зависимые столбцы. Поэтому модель без 4 столбца будет работать лучше.

```r
predictions_1 <- predict(model_1, test_data)
predictions_2 <- predict(model_2, test_data)
```

```r
print(mean((test_data['V6'] - predictions_1) ^ 2))
print(mean((test_data['V6'] - predictions_2) ^ 2))
```

Теперь построим графики остатков и посмотрим на них

```r
plot((test_data['V6'] - predictions_1)$V6)
```

```r
plot((test_data['V6'] - predictions_2)$V6)
```

Оказалось, в данных есть выброс. В целом MSE не такое плохое.
